## Functions

### printResult():
 * finds the **result** by using dot product
 * saves **result** to result.csv

```python
def printResult(mvalue):
    result = np.zeros(test_data.shape[0]).reshape(-1, 1)
    for i in range(test_data.shape[0]):
        result[i] = np.dot(test_data[i], mvalue)
    np.savetxt("result.csv", result, delimiter=",", fmt = '%1.5f')
```

### cost():
 * calculates the total cost 

```python
def cost(m):
    totalcost = 0
    N = x_train.shape[0]
    for i in range(N):
        row = x_train[i, :]
        totalcost += (1/N)*(((y_train[i][0] - np.dot(row, m)).sum())**2)
    return totalcost
```

```python
def step_gradient(learning_rate, m):
    mslope = np.zeros(x_train.shape[1]).reshape(-1, 1)
    N = x_train.shape[0]
    for i in range(x_train.shape[0]):
        row = x_train[i, :]
        for j in range(x_train.shape[1]):
            mslope[j] += (-2/N)*(((y_train[i][0] - np.dot(row, m)).sum())*row[j])
                                
    return m - learning_rate*mslope
```

```python
def gd(num_iterations, learning_rate):
    m = np.zeros(x_train.shape[1]).reshape(-1, 1)
    for i in range(num_iterations):
        m = step_gradient(learning_rate, m)
        print(i, " : Cost : ", cost(m))
    return m
```
